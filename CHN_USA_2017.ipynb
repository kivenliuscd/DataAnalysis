```python
country_media = country_media[~country_media['国家'].isin([' ','-'])]
country_media_count = country_media['媒体'].groupby(country_media['国家']).count()
country_media_count.head()
```

```python
country_media_count.ix[x3]
```

```python
country_media_count = country_media_count.reset_index(inplace=0)
country_media_count.head()
```

```python
country_media_count['媒体占比'] = country_media_count['媒体总数']/country_media_count['媒体总数'].sum()
```

```python
country_media_count['媒体总数'].sum()
```

```python
country_media_count.head()
```

```python
media_compare = country_media_count[['媒体总数','媒体占比']].groupby(country_media_count['国家']).sum().ix[x3]
media_compare = media_compare.reset_index(inplace=0)
media_compare
```

```python
country_meidia_report = data_country['媒体'].groupby([data_country['媒体国家'],data_country['媒体']]).count().ix[x3]
#媒体报道量统计
country_meidia_report = country_meidia_report.reset_index(inplace=0)
# country_meidia_report
#国家报道媒体数量
country_meidia_report = country_meidia_report['媒体'].groupby(country_meidia_report['媒体国家']).count()
country_meidia_report
```

```python
media_report_count = len(np.unique(data_country['媒体']))
country_meidia_report_persent = country_meidia_report/media_report_count
country_meidia_report_persent
```

```python
media_compare['报道媒体数量'] = country_meidia_report.values
media_compare['报道媒体占比'] = country_meidia_report_persent.values
media_compare
```

```python
avg_report = y3/media_compare['媒体总数']
avg_report
```

```python
media_compare['报道总量'] = y3
media_compare['平均报道量'] = avg_report
media_compare.to_csv(u'{}媒体归一化统计表'.format(name_head))
media_compare
```

```python
media_compare_pic1 = media_compare[['媒体占比','报道媒体占比']].groupby(media_compare['媒体国家']).sum()
y4_1 = media_compare_pic1['媒体占比']
y4_2 = media_compare_pic1['报道媒体占比']
media_compare_pic1
```

```python
normed_media_compare_pic2 = media_compare_pic1.div(media_compare_pic1.sum(1), axis=0)
y5_1 = normed_media_compare_pic2['媒体占比']
y5_2 = normed_media_compare_pic2['报道媒体占比']
normed_media_compare_pic2
```
