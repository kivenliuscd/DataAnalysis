```python
country_media_all = pd.read_csv('media_country_all.csv',header=None)
country_media_all.head()
```

```python
country_media_columns=['媒体','国家','国家1','排名']
country_media_all.rename(columns=dict(zip(np.arange(4),country_media_columns)),inplace=True)
country_media_all = country_media_all[~country_media_all['国家'].isin([' ','-'])]
country_media_all = country_media_all[~country_media_all['排名'].isin([' ','-'])][['媒体','国家','排名']]
country_media_all['排名'] = [int(i.replace(',','')) for i in country_media_all['排名']]
country_media_all.head()
```

```python
ix_country = country_media_count_heat['国家'][:10]
def top(df, n=5, column='排名'):
    return df.sort_values('排名')[:n]
```

```python
#生成Top排行榜国家媒体列表
list_country = []
for i,name in zip([50]+[30]*6+[20]*6+[10]*7,country_media_count_heat['国家']):
#     print i,name
    list_country.append(country_media_all[['媒体','排名']].groupby(country_media_all['国家']).apply(top,n=i).ix[[name]].reset_index(inplace=0)[['国家','媒体','排名']])
country_top_media = pd.concat(list_country)
# print country_top_media
f = open('country_top_media.csv','ab+')
writer = csv.writer(f)
for row in country_top_media.values:
#     print row
    writer.writerow(row)
f.close()
```

```python
#对日期进行分组
def group_dict(df, column='媒体'):
#     print df
    new = df.groupby(['日期','媒体']).count().unstack()
    new = new.groupby(date_dict,axis=0).sum().stack().reset_index(inplace=0).groupby('level_0')['媒体'].count()
#     print new
    return new
#     return df['媒体国家'][:3]
```

```python
def merge_media(df):
    new = df.reset_index(inplace=0)
    media_list = new['媒体'].tolist()
    media_list
    return '; '.join(media_list)
```

```python
country_top_media_merge = country_data.groupby('国家').apply(merge_media)
country_top_media_merge
```
